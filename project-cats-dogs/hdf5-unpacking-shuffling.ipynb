```python
import h5py
import numpy as np
from matplotlib import pyplot as plt
from skimage import data, io, filters, util, color
from skimage.transform import rescale, resize, downscale_local_mean
from skimage import img_as_ubyte
from sklearn.utils import shuffle 

import os
```

```python
with h5py.File('cat-dog-dataset.hdf5','r') as f:
    train_0 = f["train_0"]
    train_1 = f["train_1"]
    test_0 = f["test_0"]
    test_1 = f["test_1"]
    
    y_train_unsh = np.zeros((train_0.shape[0] + train_1.shape[0]), dtype=np.uint8)
    y_train_unsh[train_1.shape[0]:] = 1

    y_test_unsh  = np.zeros((test_0.shape[0] + test_1.shape[0]), dtype=np.uint8)
    y_test_unsh[test_1.shape[0]:] = 1    

    X_train_unsh = np.concatenate((train_0, train_1), axis=0)
    X_test_unsh = np.concatenate((test_0, test_1), axis=0)    
```

```python
io.imshow(X_test_unsh[341])
```

```python
X_train, y_train = shuffle(X_train_unsh, y_train_unsh) 
X_test, y_test = shuffle(X_test_unsh, y_test_unsh) 
```

```python
io.imshow(X_train[1112])
```

```python
with h5py.File('c-d-ds.hdf5','w') as f:
    dest = f.create_dataset('X_train', data=X_train)
    dest = f.create_dataset('y_train', data=y_train)
    dest = f.create_dataset('X_test', data=X_test)
    dest = f.create_dataset('y_test', data=y_test)    
```
